```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import WordCloud, get_single_color_func
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('modeled_data/word_frequencies.csv')
```

```python
df
```

```python
df_de = df[df['language'] == 'de'].copy()
df_en = df[df['language'] == 'en'].copy()
df_es = df[df['language'] == 'es'].copy()
```

```python
def fre_graph(df):

    # Assuming you have a DataFrame called 'df' with columns 'words' and 'frequency'
    word_counts = df.set_index('Word')['Frequency'].to_dict()

    wordcloud = WordCloud(width=800, height=400, background_color='black').generate_from_frequencies(word_counts)

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.axis('off')
    plt.show()
    return
```

```python
fre_graph(df_de)
fre_graph(df_en)
fre_graph(df_es)
```
