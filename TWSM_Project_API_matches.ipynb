```python
''' possible status:
SCHEDULED: Matches that have not yet been played and are scheduled for a future date and time.
LIVE: Matches that are currently being played.
IN_PLAY: Matches that are currently being played.
PAUSED: Matches that have been temporarily paused.
FINISHED: Matches that have finished.
POSTPONED: Matches that have been postponed and have not yet been rescheduled.
SUSPENDED: Matches that have been temporarily suspended and may be resumed at a later date.
CANCELED: Matches that have been canceled and will not be rescheduled.
'''
```

```python
matches_list = []
for match in response_json['matches']:
    home_team = match['homeTeam']['name']
    away_team = match['awayTeam']['name']
    if home_team == 'Bayer 04 Leverkusen' or away_team == 'Bayer 04 Leverkusen':
        match_dict = {'date': match['utcDate'],
                      'home_team': home_team,
                      'away_team': away_team,
                      'winner': match['score']['winner']}
        matches_list.append(match_dict)
```
