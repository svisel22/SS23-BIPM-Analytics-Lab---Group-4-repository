## Display the Frequency of words for each player

```python
# Load in the DataFrames for all languages
df_de = pd.read_csv('../data_files/data_clean/de_clean_condensed.csv')
df_en = pd.read_csv('../data_files/data_clean/en_clean_condensed.csv')
df_es = pd.read_csv('../data_files/data_clean/es_clean_condensed.csv')
```

```python
# Call the function from the utils which calculates word frequencies per player
df_freq_de = word_frequency_per_player(df_de, df_de['player'].unique())
df_freq_en = word_frequency_per_player(df_en, df_en['player'].unique())
df_freq_es = word_frequency_per_player(df_es, df_es['player'].unique())
```

```python
# Print frequencies per player for german data
df_freq_de = print_player_freq(df_freq_de)
```

```python
# Print frequencies per player for english data
df_freq_en = print_player_freq(df_freq_en)
```

```python
# Print frequencies per player for spanish data
df_freq_es = print_player_freq(df_freq_es)
```

```python
# Add the language column to all DataFrames 
df_freq_de['language'] = 'de'
df_freq_en['language'] = 'en'
df_freq_es['language'] = 'es'
```

```python
# Create one DataFrame which includes the word frequencies for all languages 
df_freq_all  = pd.concat([df_freq_de, df_freq_en, df_freq_es], axis=0)
```
