```python
articles = response_json['articles']

# create an empty list to hold the rows of the table
rows = []

# iterate over the articles and extract the relevant information. 
# a direct conversion into a df is not possible because 'source' is a series with two subcolumns
[{    'author': article['author'],
    'content': article['content'],
    'description': article['description'],
    'publishedAt': article['publishedAt'],
    'source_id': article['source'].get('id', ''),  # use .get() method to handle KeyError
    'source_name': article['source'].get('name', '')  # use .get()
} for article in articles]


df = pd.DataFrame(articles)

# expand the source column to separate id and name
df[['source_id', 'source_name']] = pd.json_normalize(df['source'])

# drop the original source column because otherwise we would have it and the new source_id and source_name columns (which are appended as last columns)
df.drop('source', axis=1, inplace=True)
```

# Seasonality - still working on it. It doesn't find the months correctly

```python
# Convert the 'publishedAt' column to a datetime data type
df['publishedAt'] = pd.to_datetime(df['publishedAt'])

# Create a new column that contains the month of each article's publication date
df['month'] = df['publishedAt'].dt.to_period('M')

# Group the DataFrame by the 'month' column and count the number of articles in each group
article_count = df.groupby('month').size().reset_index(name='count')

# Print the resulting table
print(article_count)
```

```python
df['month'] = df['publishedAt'].dt.month

# Group by month and count the number of articles in each group
articles_per_month = df.groupby('month')['title'].count()

# Plot the resulting dataframe
articles_per_month.plot(kind='line', x='month', y='count')

# Set the title and labels for the plot
plt.title('Number of Articles per Month')
plt.xlabel('Month')
plt.ylabel('Number of Articles')

# Show the plot
plt.show()
```
