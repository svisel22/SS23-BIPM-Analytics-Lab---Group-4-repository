```python
df_ge = df[df["language"] == "de"]
df_ge = df_ge.reset_index(drop=True)


df_en = df[df["language"] == "en"]
df_en = df_en.reset_index(drop=True)

df_es = df[df["language"] == "es"]
df_es = df_es.reset_index(drop=True)
```

```python
# Calculate the minimum length of the "data" column
min_length = df['data'].apply(len).min()
# Calculate the maximum length of the "data" column
max_length = df['data'].apply(len).max()
# Calculate the average length of the "data" column
average_length = df['data'].apply(len).mean()

print("Minimum length of the 'data' column:", min_length)
print("Maximum length of the 'data' column:", max_length)
print("Average length of the 'data' column:", average_length)
```
